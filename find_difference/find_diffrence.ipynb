```python
# GLOBAL
DOC1_PATH = u"./data/附件三_ECB政策聲明-20200123.docx"
DOC2_PATH = u"./data/附件四_ECB政策聲明-20200312.docx"
```

```python
doc1 = docx.Document(DOC1_PATH)
doc2 = docx.Document(DOC2_PATH)

class Data_analysis:
    def __init__(self, doc_path):
        self.paragraph = [i.text for i in docx.Document(doc_path).paragraphs if i.text != '']
        
        
    @property
    def fulltext(self):
        return " ".join(self.paragraph)
        
    def plot_wordcloud(self):
        pass
```

```python
from Yun_text.termMachine import termMachine
from Yun_text.dictionary import Dictionary

doc1 = Data_analysis(DOC1_PATH)
doc1_DICT = Dictionary(doc1.paragraph)
doc1_term = termMachine(doc1.fulltext)

doc2 = Data_analysis(DOC2_PATH)
doc2_DICT = Dictionary(doc2.paragraph)
doc2_term = termMachine(doc2.fulltext)

all_DICT = Dictionary([doc1.paragraph, doc2.paragraph, ['!', '?', '?', '@']])
```
